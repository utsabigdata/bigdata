```python
import numpy as np
import math
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def sigmoid(x):
    a = []
    for item in x:
        a.append(1/(1+math.exp(-item)))
    return a
```

```python
x = np.arange(-10, 10, 0.1)
sig = sigmoid(x)
```

```python
# calculate sigmoid of -6,0,6
sigmoid([-6,0,6])
```

```python
plt.plot(x,sig)
plt.show()
```
